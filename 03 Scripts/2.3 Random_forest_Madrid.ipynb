```python
# exclude MADRID from datasets for the model
df_madrid = X.loc[:, X.columns.str.startswith("MADRID")]
```

```python
# exclude MADRID from pleasent weather data
pw_madrid = y['MADRID_pleasant_weather']
```

#### converting to array and splitting the data 

```python
X = np.array(df_madrid)
y = np.array(pw_madrid)
```

```python
# Split data into train and test sets

X_train, X_test, y_train, y_test = train_test_split(X,y,random_state = 42)
```

```python
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

### Random Forest model

```python
# Create a RF classifier
clf = RandomForestClassifier(n_estimators = 100)#, max_depth=5)  
  
# Training the model on the training dataset
# fit function is used to train the model using the training sets as parameters
clf.fit(X_train, y_train)
```

```python
# Perform predictions on the test dataset
y_pred = clf.predict(X_test)
  
# Use metrics module for accuracy calculation
print("Model Accuracy: ", metrics.accuracy_score(y_test, y_pred))
```

```python
# Define class names for readability
class_names = ['Unpleasant Weather', 'Pleasant Weather']

# Plot the tree
fig = plt.figure(figsize=(60, 30))
plot_tree(clf.estimators_[15], 
          filled=True, 
          fontsize=20,
          feature_names=None,  # You can add your feature names here if you'd like
          class_names=class_names);  # Adds class labels at the leaves
```

```python
# Create a Series for MADRID feature importances
important_madrid = pd.Series(clf.feature_importances_, index=df_madrid.columns)

# Plot all features
important_madrid.sort_values(ascending=False).plot(kind='barh', figsize=(8, 5))
plt.title("MADRID Feature Importances")
plt.xlabel("Importance Score")
plt.gca().invert_yaxis()  # Show most important at the top
plt.tight_layout()
plt.savefig(r"C:\Users\kaan\ML_2\04 Analysis\Visualizations\variable_importances_madrid.png",dpi=300,bbox_inches='tight')
plt.show()
```
